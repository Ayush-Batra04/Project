```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.express as px
```

```python
df=pd.read_csv("D:\Python\Zomato Dataset\zomato.csv")
df.head()
```

```python
print(df.shape)
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
df=df.drop(['url','phone','reviews_list','menu_item'],axis=1)
df.head()
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
def clean(rate):
    if(rate=='NEW' or rate=='-'):
        return np.nan
    else:
        r=str(rate).split('/')
        r=r[0]
        return float(r)
df['rate']=df['rate'].apply(clean)
df
```

```python
print(df['rate'].mean())
df['rate']=df['rate'].fillna(3.7)
df['rate'].isna().sum()
```

Calculate Average Rating of each restaurant

```python
df_rate=df.groupby('name')['rate'].mean().to_frame()
df_rate=df_rate.reset_index()
df_rate.columns=['restaurant','rating']
df_rate.sort_values('rating',ascending=False)
df_rate.head(20).style.background_gradient(cmap="Wistia")
```

Top voted restaurant?

```python
df_vote=df.groupby('name').agg(total_votes=('votes','sum'))
df_vote=df_vote.reset_index()
df_vote=df_vote.sort_values('total_votes',ascending=False)
df_vote=df_vote[0:10]
sns.set_palette("BuGn_r", 10)
sns.catplot(x='name',y='total_votes',data=df_vote,kind='bar')
plt.xticks(rotation=90)
```

Which dish is liked most ?

```python
dish=[]
df_dish=df[df['dish_liked'].notnull()]
for index,row in df_dish.iterrows():
    dish.append(row["dish_liked"])
dish_1=[]
for i in dish:
    l=[]
    l=i.split(',')
    for j in l:
        dish_1.append(j)
dish_2=pd.DataFrame(dish_1,columns=['Dish'])
dish_2.head()
df_dishcount=dish_2['Dish'].value_counts().to_frame()
df_dishcount=df_dishcount.reset_index()
df_dishcount=df_dishcount[0:10]
sns.barplot(x='index',y='Dish',data=df_dishcount)
plt.xticks(rotation=60)
```
